```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

### Simple Linear regression
### $\hat{y} = \hat{\beta_0} + \hat{\beta_1}x$

```python
# Parameters for the data
n_points = 100
x_min, x_max = 0, 10
slope = 2  # The slope of the linear trend
intercept = 3  # The intercept of the linear trend
noise_std_dev = 1  # Standard deviation for Gaussian noise

# Generate the data
np.random.seed(0)  # For reproducibility
x = np.random.uniform(x_min, x_max, n_points)
y = slope * x + intercept + np.random.normal(0, noise_std_dev, n_points)
```

### Estimate $\hat{\beta_0}$ and $\hat{\beta_1}$ by minimizing RSS (Residual Sum of Squares)
### RSS = $\sum_{i=1}^{n}(\hat{\epsilon_i})^2 = \sum_{i=1}^{n}(y_i - (\hat{\beta_0} + \hat{\beta_1}x_i))^2$
### $\hat{\beta_1} = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$
### $\hat{\beta_0} = \bar{y} - \hat{\beta_1}\bar{x}$

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

b1_hat = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
b0_hat = y_mean - b1_hat * x_mean

y_hat = b0_hat + b1_hat * x

plt.scatter(x, y, s=20, color="blue", label="Data points")
plt.plot(x, y_hat, color="red", label="Linear Regression line")
plt.legend()
plt.grid(True, linestyle=":", color="gray")
plt.xlabel("x (data point)")
plt.ylabel(r"$\hat{y} = \hat{\beta_0} + \hat{\beta_1}x$ (prediction)")
plt.title("Simple Linear Regression")
```

```python
display(Math(r"\hat{\beta_0} = " + f"{b0_hat:.2f}"))
display(Math(r"\hat{\beta_1} = " + f"{b1_hat:.2f}"))
```
